```python
MIN_PRIME = 2
def is_prime(num):
    num = int(num)
    if(num < MIN_PRIME):
        return False
    
    lowbound = 2
    upbound = int(num / 2 + 1)
    for i in range(lowbound, upbound):
        if num % i == 0:
            return False
    
    return True

# number = input("sdfsfasf")
# if is_prime:
#    print("prime")
# else:
#    print("not prime")

assert(is_prime(3) == True)
assert(is_prime(4) == False)
assert(is_prime(7) == True)
```
